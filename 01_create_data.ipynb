```python
# drop product category and aggregate data
baci_raw.drop('k', axis=1, inplace=True)
baci = baci_raw.groupby(by=['t','i','j']).sum().reset_index()

# add exporting country names
baci = baci.merge(country_codes[['country_code', 'country_name_full', 'iso_3digit_alpha']], left_on=['i'] , right_on=['country_code'], how='left')
baci.drop('country_code', axis=1, inplace=True)

# rename columns
baci.rename(columns = {'country_name_full':'country_export', 'iso_3digit_alpha':'iso3_export'}, inplace = True)

# add importing country names
baci = baci.merge(country_codes[['country_code', 'country_name_full', 'iso_3digit_alpha']], left_on=['j'] , right_on=['country_code'], how='left')
baci.drop('country_code', axis=1, inplace=True)

# rename columns
baci.rename(columns = {'country_name_full':'country_import', 'iso_3digit_alpha':'iso3_import'}, inplace = True)


baci
```

## Weights

```python

#baci['v_normalised'] = baci['v'] / baci['v'].max()
```

```python
## Case 1: Undirected network
## produce undirected edge list and sum over import and exports
baci1 = baci.copy()

baci1['edge'] = baci.apply(lambda x: str(x['i'])+',' + str(x['j']) if x['i'] < x['j'] else str(x['j'])+','+ str(x['i']), axis=1)
baci1 = baci1.groupby(by=['t','edge']).sum().sort_values('i', ascending=False).reset_index()[['t','edge','v']]
baci1['p'] = baci1.apply(lambda x: np.int32(x['edge'].split(',')[0]), axis=1)
baci1['q'] = baci1.apply(lambda x: np.int32(x['edge'].split(',')[1]), axis=1)
baci1 = baci1.sort_values(by=['t','p','q'])
baci1.drop('edge', axis=1, inplace=True)
baci1
```

```python
baci1.to_pickle('../../data/baci_unweighted.pkl')
del baci1
```

```python
## Case 2: Directed network
baci2 = baci.copy()

## scale weights
## weights of outgoing edges are scaled as share of the total trade (import and export) per country to capture importance
g = baci2.groupby(['t', 'i'])['v']

min_, max_ = g.transform('min'), g.transform('max')
baci2['v' + '_minmax_scale'] = (baci2['v'] - min_) / (max_ - min_)

sum_ = g.transform('sum')
baci2['v' + '_share'] = baci2['v'] / sum_
```

```python
## case 3
# undirected average of directed network
baci3 = baci2.merge(baci2[['t','i', 'j', 'v_minmax_scale']], left_on=['t','i','j'] , right_on=['t','j','i'], how='left')
baci3['v_minmax_avg'] = baci3.apply(lambda x: x['v_minmax_scale_x'] if pd.isna(x['v_minmax_scale_y']) else np.mean([x['v_minmax_scale_x'],x['v_minmax_scale_y']]), axis=1)
baci3.drop(['v','q','v_minmax_scale_x','v_minmax_scale_y','i_y', 'j_y','v_share'], axis=1, inplace=True)
baci3.rename(columns = {'i_x':'i', 'j_x':'j'}, inplace = True)
```

```python
## case 4  Null model of case 3: gleichverteilte edge weights per group in directed graph, dann averaging
baci4_dummy = baci.copy()

## scale weights
## weights of outgoing edges are scaled as share of the total trade (import and export) per country to capture importance
g = baci4_dummy.groupby(['t', 'i'])['v']

mean_ = g.transform('sum')
baci4_dummy['v' + '_mean_scale'] = mean_ 

display(baci4_dummy)

# undirected average of directed network
baci4 = baci4_dummy.merge(baci4_dummy[['t','i', 'j', 'v_mean_scale']], left_on=['t','i','j'] , right_on=['t','j','i'], how='left')
del baci4_dummy
baci4['v_mean_scale_avg'] = baci4.apply(lambda x: x['v_mean_scale_x'] if pd.isna(x['v_mean_scale_y']) else np.mean([x['v_mean_scale_x'],x['v_mean_scale_y']]), axis=1)
baci4.drop(['v','q','v_mean_scale_x','v_mean_scale_y','i_y', 'j_y'], axis=1, inplace=True)
baci4.rename(columns = {'i_x':'i', 'j_x':'j'}, inplace = True)
```
